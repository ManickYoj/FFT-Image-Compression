```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy
import pandas
import scipy.signal
from scipy.sparse import coo_matrix

from IPython.display import Image, display
from matplotlib import pyplot

pyplot.rcParams['image.cmap'] = 'gray'
pyplot.rcParams['image.interpolation'] = 'nearest'
```

```python
def plot_image(image):
    pyplot.imshow(image)
    thinkplot.Show(legend=False)
```

```python
import matplotlib.image
image = matplotlib.image.imread('otter.jpg')
plot_image(image)
```

```python
def roll_spectrum(spectrum):
    n, m = spectrum.shape
    spectrum = numpy.roll(spectrum, n//2, axis=0)
    spectrum = numpy.roll(spectrum, m//2, axis=1)
    return spectrum
```

```python
def plot_spectrum(spectrum):
    roll_spectrum(spectrum)
    plot_image(numpy.log(numpy.absolute(spectrum)))
```

Visualizing the FT of the Image's Greyscale.

```python
image_grey = image.sum(axis=2)
plot_image(image_grey)
image_grey_fft = numpy.fft.rfft2(image_grey)
plot_spectrum(image_grey_fft)
```

```python
def recover_image(image_fft):
    image_recovered = numpy.fft.irfft2(image_fft)
    plot_image(image_recovered.real)
```

```python
def normalize(array):
    """
    Given an array with a range of values, + & -,
    normalizes all values to between -1 and 1.
    """
    return array / numpy.max(numpy.abs(array))

# Test that normalization works with irfft2
recover_image(image_grey_fft)
recover_image(normalize(image_grey_fft))
```

###The Fun Begins
The below code simulates the visual effects of image compression by zeroing out all components with amplitude lower
than the designated cutoff. Compare the image above this to the image outputted below. At an amplitude cutoff of 3e-4,
(higher = more compression/lossier) the image is "compressed" to 1/20 of the size with noticiable but mild quality degredation.

Feel free to play with the cutoff amplitude to view its effect.

```python
def previewCompression(image_fft, cutoff_amplitude=3e-4, verbose=False):
    # 0 out all components below the cutoff
    normed_fft = normalize(image_fft)
    numpy.place(normed_fft, abs(normed_fft) < cutoff_amplitude, [0])
    
    if verbose:
        recover_image(normed_fft)  # Display compressed image
        
        # Display information on the compression of the image
        original_size = normed_fft.shape[0]*normed_fft.shape[1]
        components_removed = numpy.argwhere(abs(normed_fft) < cutoff_amplitude).shape[0]
        percent_removed = (1-components_removed/original_size)*100
        print("Removed {} out of {} components. Image contains {}% of the original components.".format(components_removed, original_size, percent_removed))
    
    return normed_fft
```

```python
image = matplotlib.image.imread('otter.jpg')
image_grey = image.sum(axis=2)
image_numpy_fft = numpy.fft.rfft2(image_grey)
image_scipy_fft = scipy.fftpack.fft2(image_grey)

image_compressed_fft_numpy = previewCompression(image_numpy_fft, verbose=False)
image_compressed_fft_scipy = previewCompression(image_scipy_fft, verbose=False)
```

```python
def compress(image, cutoff=3e-4):
    """
    Takes a color or greyscale image and compresses it
    by representing it as a sparse matrix with high frequency components removed.
    This is a lossy compression technique. Lower cutoffs lead to a smaller file
    of lesser quality. The returned image will be greyscale regardless of input.
    
    Args:
      image - the image to be compressed
    """
    # Handle color images (Currently bugged)
    if image.ndim == 3:  # If in color (or an RGB layer which believes its 3d)
        if image.shape[2] == 3:  # If a color image, run layers through compress seperately, and bind together the result
            return [compress(layer) for layer in numpy.dsplit(image, 3)]
        else:  # If an RGB layer, convert 3d (with 3rd layer of 1 depth) to 2d        
            image = numpy.reshape(image, image.shape[0:2])

    # Handle greyscale or single layer images
    image_fft = numpy.fft.rfft2(image)  # Take RFFT
    normed_fft = image_fft / numpy.max(numpy.abs(image_fft))  # Normalize values
    numpy.place(normed_fft, abs(normed_fft) < cutoff, [0])  # Replace values below cutoff with 0
    return [coo_matrix(normed_fft)]  # Format as sparse matrix

def decompress(compressed):
    channels = [numpy.fft.irfft2( layer.todense() ) for layer in compressed]
    if len(channels) == 1: return channels[0]
    else: return numpy.dstack(channels)
```
