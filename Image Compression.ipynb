### Setting up

Below are a variety of small utility functions, some taken from the Comp DSP lectures, and others written here which were used in creating, testing, or presenting the below workbook. The code below also performs an initial import of an image of an otter. This image is a good candidate for testing lossy compression because it has some visible high-frequency components (fur) on the snout which makes high-frequency filtering apparent, as well as a variety of low frequency components in the background for comparison.

```python
def plot_image(image):
    pyplot.imshow(image)
    thinkplot.Show(legend=False)
    
def roll_spectrum(spectrum):
    n, m = spectrum.shape
    spectrum = numpy.roll(spectrum, n//2, axis=0)
    spectrum = numpy.roll(spectrum, m//2, axis=1)
    return spectrum

def plot_spectrum(spectrum):
    roll_spectrum(spectrum)
    plot_image(numpy.log(numpy.absolute(spectrum)))
    
def recover_image(image_fft):
    image_recovered = numpy.fft.irfft2(image_fft)
    plot_image(image_recovered.real)
    
def normalize(array):
    """
    Given an array with a range of values, + & -,
    normalizes all values to between -1 and 1.
    """
    return array / numpy.max(numpy.abs(array))

import matplotlib.image
image = matplotlib.image.imread('otter.jpg')
image_grey = image.sum(axis=2)
image_grey_fft = numpy.fft.rfft2(image_grey)

plot_image(image)
plot_image(image_grey)
plot_spectrum(image_grey_fft)
```

###The Fun Begins

The below code simulates the visual effects of the image compression technique to be used by zeroing out all components with amplitude lower than the designated cutoff. Compare the two images outputted below. At an amplitude cutoff of 1e5, (higher = more compression/lossier) the image is "compressed" to a a little over 6% of the original size with noticiable but acceptable quality degredation.

A note - the cutoff amplitude has a rather arbitrary relation to different images. Originally, frequency component amplitudes were normalized to the range -1 to 1 using the normalize function shown in setup. However, when attempting to work with color images, it became apparent that, because the RGB layers are treated seperately, normalization destroyed the original relation between their values and made the decompressed image useless. Therefore, normalization is no longer a part of the compression or previewCompression process. This has no visual effect on the result, but it does mean that the cutoff amplitudes no longer fall in a well defined -1 to 1 range.

```python
def previewCompression(input_image_fft, cutoff_amplitude=1e5, verbose=False):
    # 0 out all components below the cutoff
    output_image_fft = input_image_fft.copy()
    numpy.place(output_image_fft, abs(output_image_fft) < cutoff_amplitude, [0])
    
    if verbose: 
        recover_image(output_image_fft)  # Display compressed image
        
        # Display information on the compression of the image
        original_size = output_image_fft.shape[0]*output_image_fft.shape[1]
        components_removed = numpy.argwhere(abs(output_image_fft) < cutoff_amplitude).shape[0]
        percent_removed = (1-components_removed/original_size)*100
        print("Removed {} out of {} components. Image contains {}% of the original components.".format(components_removed, original_size, percent_removed))
    
    return output_image_fft

print("Uncompressed Image")
plot_image(image_grey)

print("Compressed Image")
image_compressed_fft = previewCompression(image_grey_fft, verbose=True)
```

### Interactive Compression Preview Widget

Below is a widget that allows the reader to play with various compression thresholds to see their effect on the image. Left gives higher quality, right gives higher compression. Notice that a percentage of removed components is given below the image.

```python
from IPython.html import widgets

image = matplotlib.image.imread('otter.jpg')
image_grey = image.sum(axis=2)
input_image_fft = numpy.fft.rfft2(image_grey)

def previewWidget(**kwargs):
    previewCompression(input_image_fft, kwargs['cutoff_amplitude'], True)

widgets.interact(previewWidget, cutoff_amplitude=(1, 8e6))
```

### A Quick Aside: numpy RFFT vs scipy FFT

If you were reading closely above, you will notice that I opted to use the numpy.fft module instead of the scipy.fftpack module to take the fft of the image. I moved over to numpy because scipy's fftpack does not provide a real-fft or inverse-real-fft for 2d data. While its fft2 and ifft2 still work (if you pull out just the real components on the ifft2) they double the number of necessary bytes taken up by the image, as shown below. As the point of my project was image compression, an unnecessary doubling of the fft's data size seemed unnecessary.

```python
image_compressed_fft_scipy = previewCompression(scipy.fftpack.fft2(image_grey));
image_compressed_fft_numpy = previewCompression(numpy.fft.rfft2(image_grey));

print("Original: " +  str(image_grey.nbytes))

print("\n--- Numpy Bytesizes After RFFT2 and IRFFT2 ---")
print("Image after rfft: " + str(numpy.fft.rfft2(image_grey).nbytes))
print("Image after irfft(rfft): " + str(numpy.fft.irfft2(numpy.fft.rfft2(image_grey)).nbytes))

print("\n---Numpy Bytesizes after Compression Preview ---")
print("Image after compressionPreview(rfft): " + str(image_compressed_fft_numpy.nbytes))
print("Image after irrft(compressionPreview(rfft)): " + str(numpy.fft.irfft2(image_compressed_fft_numpy).nbytes))

print("\n--- Scipy Image Bytesizes After FFT2 and RFFT2 ---")
print("Image after fft2: " + str(scipy.fftpack.fft2(image_grey).nbytes))
print("Image after ifft2(fft2): " + str(scipy.fftpack.ifft2(scipy.fftpack.fft2(image_grey)).nbytes))

print("\n--- Scipy Image Bytesizes after Compression Preview ---")
print("Image after compressionPreview(fft2): " + str(image_compressed_fft_scipy.nbytes))
print("Image after ifft2(compressionPreview(fft2)): " + str(scipy.fftpack.ifft2(image_compressed_fft_scipy).nbytes))
```

### Compression test

Compressing the fft is straightforward - after culling all of the frequency components below the cutoff value, we convert the numpy array into a coordinate sparse matrix (scipy.sparse.coo_matrix), which enters all of the nonzero components of the array into a (x, y) -> value dictonary.

In order to get a sense for whether or not dropped components correspond well to the image's compression ratio, we need save the files to disk. For simplicity, we use the pickle filetype provided by the cPickle module. An important note: pickle files are convenient, but certainly not compact. The original jpeg image of the otter is approximately 151 kB - the pickled version of the same data loaded into a numpy array is 8,403 kB. Accepting that the baseline file sizes are much larger (15,346 kB for the baseline greyscale image), then it is still worth noting that, although taking the fft of the image almost doubles the file size (to 24,275kB), using the compression as presented above reduces the file size down to 2,447 kB: about 1/6 of the baseline.

Note: Notices that the file size of the color image is smaller (almost by a factor of 2) than the greyscale file size. We did not satisfactorily conclude the reason for this, but it may be related to a transition from a uint8 datatype for the color image to a uint32 datatype for the greyscale image.

```python
import cPickle
import os.path

image = matplotlib.image.imread('otter.jpg')
image_grey = image.sum(axis=2)
image_compressed_fft = previewCompression(image_grey)

# Curious about dtypes
# print(image.dtype)  # uint 8
# print(image_grey.dtype)  # uint 32
# print(image_compressed_fft.dtype)  # uint 32

# Designate image paths
uncompressed_color_img_path = os.path.join('saves','uncompressedimgcolor.pkl')
uncompressed_grey_img_path = os.path.join('saves','uncompressedimg.pkl')
uncompressed_fft_path = os.path.join('saves','uncompressedfft.pkl')
compressed_fft_path = os.path.join('saves','compressedfft.pkl')

# Save images to compare file sizes
cPickle.dump(image, open(uncompressed_color_img_path, 'wb'))
cPickle.dump(image_grey, open(uncompressed_grey_img_path, 'wb'))
cPickle.dump(numpy.fft.rfft2(image_grey), open(uncompressed_fft_path, 'wb'))
cPickle.dump(coo_matrix(image_compressed_fft), open(compressed_fft_path, 'wb'))

# Display file sizes
print("Uncompressed color image size: " + str(os.path.getsize(uncompressed_color_img_path)))
print("Uncompressed greyscale image size: " + str(os.path.getsize(uncompressed_grey_img_path)))
print("Uncompressed greyscale fft size: " + str(os.path.getsize(uncompressed_fft_path)))
print("Compressed greyscale fft size: " + str(os.path.getsize(compressed_fft_path)))

# Load back all images to ensure data was preserved.
recover_image(cPickle.load(open(uncompressed_fft_path, 'rb')))
recover_image(cPickle.load(open(compressed_fft_path, 'rb')).todense())
```

### Condensation of Compression Code

For the compression tests, we strung together a variety of commands to compress an image. This is a rather clumsy approach. Therefore, functions are provided below to simplify the use of the compression and decompression algorithm. Currently, compression is only supported for greyscale images. A test of the compress/decompress functions and a comparison of compressed filesizes including a compression ratio display follows the function definitions.

Color images are currently able to be compressed and decompressed, with their RGB channels appearing to go though the process intact. However, compositing the image fails for reasons yet unknown, but likely related to the data-type of the numpy array. A note - if the component culling step is removed, image compositing seems to work, but lowering the culling cutoff to 0 seems to have no effect.

```python
def compress(image, cutoff=1e5):
    """
    Takes a greyscale image and compresses it into a sparse matrix
    with low amplitude components removed. Decompression can be achieved
    using the decompress function.
    
    Args:
        - image: The greyscale image, represented as a 2d numpy array, to be compressed
        - cutoff: The amplitude cutoff used to judge whether to discard frequency components. 
                  A low cutoff will lead to a higher quality image but with a larger filesize.
                  
    Returns:
        - The compressed image, which can be restored using the decompress function.
    """
    # Handle color images (Currently bugged)
    if image.ndim == 3:  # If in color (or an RGB layer which believes its 3d)
        if image.shape[2] == 3:  # If a color image, run layers through compress seperately, and bind together the result
            return [compress(layer) for layer in numpy.dsplit(image, 3)]
        else:  # If an RGB layer, convert 3d (with 3rd layer of 1 depth) to 2d        
            image = numpy.reshape(image, image.shape[0:2])

    # Handle greyscale or single layer images
    image_fft = numpy.fft.rfft2(image)  # Take RFFT
    # Note: commenting the following line destroys the "compression" aspect, but allows for compression/decompression
    # to work on color images.
    numpy.place(image_fft, abs(image_fft) < cutoff, [0])  # Replace values below cutoff with 0
    compressed = coo_matrix(image_fft)  # Format as sparse matrix
    return compressed


def decompress(compressed):
    """
    Takes an image compressed by the compress function and decompresses it to a plotable
    numpy array.
    
    Args:
        - compressed: The compressed image
        
    Returns:
        - A numpy array ready for plotting.
    """
    if isinstance(compressed, list): # Handle color images (list with 3 coo matricies)
        rgb = numpy.dstack([numpy.fft.irfft2( layer.todense() ) for layer in compressed])
        rgb_array = numpy.array(rgb, dtype='uint8')
        return rgb_array
    else: # Handle greyscale (1 coo matrix)
       return numpy.fft.irfft2(compressed.todense())
```

```python
image = matplotlib.image.imread('otter.jpg')
image_grey = image.sum(axis=2)
plot_image(decompress(compress(image_grey)))
#plot_image(decompress(compress(image, 0)))  # Color attempts - they are currently bugged
```
